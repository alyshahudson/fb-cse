# Reads in N-Grams

```python
import pandas as pd
import re

rules_df = pd.read_csv('N-Gram Analysis.csv')
rules_df = rules_df.drop(rules_df.columns[0], axis = 1)
```

```python
rules_df
```

# Reads in Rules

```python
r_analysis_df = pd.read_csv('Rule Analysis.csv')
r_analysis_df = r_analysis_df.drop(r_analysis_df.columns[0], axis = 1)
```

```python
r_analysis_df
```

# Consolidates N-Grams and Rules

```python
for i in range(len(r_analysis_df)):
    term = r_analysis_df['Terms'][i]
    exception = r_analysis_df['Exceptions'][i]
    aff = r_analysis_df['Affiliation'][i]
    rep = r_analysis_df['Republican Usage'][i]
    dem = r_analysis_df['Democrat Usage'][i]
    tot = r_analysis_df['Total Usage'][i]
    acc = r_analysis_df['Accuracy %'][i]
    
    entry = {'Terms': term, 'Exceptions': exception, 'Republican Usage': rep, 'Democrat Usage': dem, 'Total Usage': tot, 'Affiliation': aff, 'Accuracy %': acc}
    
    rules_df = rules_df.append(entry, ignore_index = True)
```

```python
rules_df
```

# Reads in Voter Data

```python
users_df = pd.read_csv('Unlabeled Data.csv')

#removes unwanted information and standardizes messages -- lowercase, removes additional symbols, etc.
users_df = users_df.drop(users_df.columns[0], axis = 1)
users_df['Message'] = users_df['Message'].apply(lambda x: x.lower())
users_df['Message'] = users_df['Message'].apply(lambda x: x.replace('nan', ''))
users_df['Message'] = users_df['Message'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))
```

```python
users_df
```

# Set Model Paramaters

```python
# rule accuracy -- only rules with > accuracy %  will be used
rule_cutoff = 70

# classification certainty -- only voters with > accuracy %  will be categorized
certainty_cutoff = 75
```

# Create DataFrame for Rule Stats

```python
data = [rules_df['Terms'], rules_df['Exceptions'], rules_df['Affiliation'], rules_df['Accuracy %']]
headers = ['Terms', 'Exceptions', 'Affiliation', 'Accuracy %']

rule_stats = pd.concat(data, axis = 1, keys = headers)
```

```python
rule_stats
```

```python
#counts frequency of rule in unlabeled data
def freq(x):
    rule = x['Terms']
    sum_freq = 0
    
    #makes exceptions lowercase -- standardizes
    try:
        exceptions = x['Exceptions'].lower().split(",")
    except:
        exceptions = None
        
        
    found_rules = list(users_df[users_df['Message'].str.contains(rule)].index)
    
    
    #iterates through all indices of entries containing rule
    for i in found_rules:
        message = users_df['Message'][i]
        tot_rule = message.count(rule)
        tot = 0
         
        # case 1 -- rule has exceptions
        if exceptions:
            tot_exception = 0
            
            for excep in exceptions:
                exception_i = message.count(excep)
                tot_exception += exception_i
                
            tot = tot_rule - tot_exception
          
        # case 2 -- all other rules
        else:
            tot = tot_rule
                
        sum_freq += tot

        
    return sum_freq    
```

```python
rule_stats['Total Usage'] = rule_stats.apply(freq, axis = 1)
```

```python
rule_stats = rule_stats.sort_values('Total Usage', ascending = False)
```

```python
rule_stats
```

# Counts Unlabeled Voters' Rule Usages

#### iterates through each voters' comments and counts Republican and Democrat associated rules

```python
import math

def count(x, party_aff, rule_cutoff):
    message = x['Message']
    rule_cutoff = rule_cutoff / 100
    score = 0
    
    #filters rules based on party affiliation being counted
    all_aff_rules = rules_df[rules_df['Affiliation'] == party_aff]
    all_aff_rules_i = list(all_aff_rules['Terms'].index)

    
    #iterates through corresponding rules
    for i in all_aff_rules_i:
        rule = all_aff_rules['Terms'][i]
        accuracy = all_aff_rules['Accuracy %'][i] / 100
        tot = 0
        
        #filters out rules having less than accuracy cutoff
        if accuracy > rule_cutoff and not math.isnan(accuracy):
            tot_rule = message.count(rule)

        
        # case 1 -- rule has exceptions
            try:
                exceptions = all_aff_rules['Exceptions'][i].split(",")
                tot_exception = 0

                for excep in exceptions:
                    exception_i = int(message.count(excep))
                    tot_exception += exception_i

                #ignores exceptions
                tot = tot_rule - tot_exception
            
            # case 2 -- all other rules
            except:
                tot = tot_rule

                
            #rules' score measured by its accuracy
            score = score + (tot * accuracy)
  
    return score
```

```python
users_df['Republican Count'] = users_df.apply(count, args = (0, rule_cutoff), axis = 1)
users_df['Democrat Count'] = users_df.apply(count, args = (1, rule_cutoff), axis = 1)
```

```python
users_df
```

# Determines User's Party Affiliation and Calculates % Certainty

#### compares each users' Republican and Democrat rule count

```python
def certainty(x):
    r_count = x['Republican Count']
    d_count = x['Democrat Count']
    sum_count = r_count + d_count
    
    if sum_count > 0:
        if r_count > d_count:
            tot = (r_count / sum_count) * 100
        else:
            tot = (d_count / sum_count) * 100

        tot = float('{0:.2f}'.format(tot))
        
        return tot

    else:
        return None
 
```

```python
def categorization(x, certainty_cutoff):
    r_count = x['Republican Count']
    d_count = x['Democrat Count']
    certainty = x['% Certainty']
    
    if certainty >= certainty_cutoff:
        if r_count > d_count:
            return 0
        else:
            return 1
    else:
        return None
    
```

```python
users_df['% Certainty'] = users_df.apply(certainty, axis = 1)
```

```python
users_df['Categorization'] = users_df.apply(categorization, args = (certainty_cutoff, ), axis = 1)
```

```python
users_df
```

# Voter Data Analysis

```python
#counts the total number of users -- ID'ed and not
users_id = users_df[users_df['Categorization'] >= 0]

tot_users_id = len(users_id)
tot_users = len(users_df)
```

```python
#counts total Reps and Dems ID'ed
reps_id = users_id[users_id['Categorization'] == 0]
dems_id = users_id[users_id['Categorization'] == 1]

tot_reps_id = len(users_id[users_id['Categorization'] == 0])
tot_dems_id = len(users_id[users_id['Categorization'] == 1])
```

```python
#dataframe for voter data analysis
data = {'Groups' : ['Users', 'Users IDed', 'Republicans IDed', 'Democrats IDed'], 
        'Totals' : [tot_users, tot_users_id, tot_reps_id, tot_dems_id]}

user_data = pd.DataFrame(data, columns = ['Groups', 'Totals'])
```

```python
user_data
```

# Visual Represenations of Data

```python
from pandas import Series
import seaborn as sns
import matplotlib.pyplot as plt

#values to be ploted
labels = Series(user_data['Groups'])
values = Series(user_data['Totals'])
descending = user_data.sort_values('Totals', ascending = False).Groups

plt.figure(figsize = (13,5))
sns.set(font_scale = 1.2)
sns.barplot(x = values, y = labels, data = user_data, order = descending, palette = 'Oranges_r').set_title('VCM Voter Data Analysis')
```

```python
#creates figure with 2 subplots (donut pie charts)
fig, (plot_id, plot_aff) = plt.subplots(1, 2, figsize = (18,7))

fig.suptitle('VCM Voter Analysis', fontsize = 'xx-large', ha = 'center')
txt = ('Rule Certainty: ' + str(rule_cutoff) + '% \n' +
         'Model Categorization Certainty: ' + str(certainty_cutoff) + '%')
fig.text(0, 0, txt, fontsize = 'large')
fig.subplots_adjust(left = 0.05, bottom = 0.1)

# circles for donut pie chart
donut = plt.Circle( (0,0), 0.6, color = 'White')
donut_2 = plt.Circle( (0,0), 0.6, color = 'White')


# 1 -- % Voters IDed
labels_users_id = ['IDed \n (' + str(tot_users_id) + ')', 'Total \n (' + str(tot_users) + ')']
size_users_id = [tot_users_id, tot_users]
per = '{0:.2f}'.format((tot_users_id / tot_users) * 100)

colors_users_id = ['bisque', 'orange']
plot_id.pie(size_users_id, labels = labels_users_id, colors = colors_users_id)
plot_id.set_title('% Voters IDed', fontsize = 'x-large')
plot_id.add_artist(donut)
plot_id.text(x= -.2, y = 0, s = str(per) + "%", fontsize = 'x-large')


# 2 -- % Democrats IDed
labels_users_aff = ['Republicans \n (' + str(tot_reps_id) + ')', 'Democrats \n (' + str(tot_dems_id) + ')']
size_users_aff = [tot_reps_id, tot_dems_id]
per = (tot_dems_id / tot_users_id) * 100
per = '{0:.2f}'.format(per)

colors_users_aff = ['salmon', 'cornflowerblue']
plot_aff.pie(size_users_aff, labels = labels_users_aff, colors = colors_users_aff)
plot_aff.set_title('% Democrats IDed', fontsize = 'x-large')
plot_aff.add_artist(donut_2)
plot_aff.text(x= -.2, y = 0, s = str(per) + "%", fontsize = 'x-large')
```
