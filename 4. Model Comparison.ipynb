# Reads in Model Input Files

#### Clelia's Voter Categorization Model (VCM)

```python
import pandas as pd
import re

vcm_model = pd.read_csv('VCM Categorization Data.csv')
vcm_model = vcm_model.drop(vcm_model.columns[0], axis = 1)
vcm_model['Message'] = vcm_model['Message'].apply(lambda x: str(x).replace('nan', '') if (str(x) == 'nan') else x)
```

```python
vcm_model
```

#### Rupali's ML Model

```python
ml_model = pd.read_csv('Rupali ML Data.csv')
ml_model = ml_model.drop(ml_model.columns[0], axis = 1)

#standardizes data -- lowercase, removes all whitespaces
ml_model['Message'] = ml_model['Message'].apply(lambda x: x.lower())
ml_model['Message'] = ml_model['Message'].apply(lambda x: x.replace('nan', ''))
ml_model['Message'] = ml_model['Message'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))
```

```python
#categorizes voters based off % dem value
def ml_cate(x):
    per_dem = x['Percent_Dem']
    
    #voters with 50% dem are not categorized
    if per_dem > .5:
        return 1
    elif per_dem < .5:
        return 0
    else:
        return None
```

```python
ml_model['Categorization'] = ml_model.apply(ml_cate, axis = 1)
```

```python
ml_model
```

# Create DataFrame to Store Comparison Data

```python
data = [ml_model['Name'], ml_model['Message'], ml_model['Percent_Dem'], ml_model['Categorization'], vcm_model['Categorization'], vcm_model['% Certainty']]
headers = ['User', 'Message', 'ML Model -- % Dem', 'ML Model -- Categorization', 'Voter Cat Model -- Categorization', 'Voter Cat Model -- % Certainty']

comp_df = pd.concat(data, axis = 1, keys = headers)
```

```python
import math

def valid(x):
    ml = x['ML Model -- Categorization']
    vcm = x['Voter Cat Model -- Categorization']
    
    if not math.isnan(vcm):
        if ml == vcm:
            return True
        else:
            return False
    else:
        return None
```

```python
comp_df['Comparison'] = comp_df.apply(valid, axis = 1)
```

```python
comp_df
```

```python
comp_users = comp_df[comp_df['Comparison'] >= 0]
validated_users = comp_df[comp_df['Comparison'] == True]
```

```python
#all voters both models labeled the same --> validated
validated_users
```

# Models' Data Analysis

```python
#counts total users -- ID'ed and not
ml_id = ml_model[ml_model['Categorization'] >= 0]
vcm_id = vcm_model[vcm_model['Categorization'] >= 0]

tot_users = len(comp_df)
tot_ml_id = len(ml_id)
tot_vcm_id = len(vcm_id)
```

```python
#counts totals for ML model
ml_reps_id = ml_model[ml_model['Categorization'] == 0]
ml_dems_id = ml_model[ml_model['Categorization'] == 1]

tot_ml_reps = len(ml_reps_id)
tot_ml_dems = len(ml_dems_id)
```

```python
#counts totals for VCM model
vcm_reps_id = vcm_model[vcm_model['Categorization'] == 0]
vcm_dems_id = vcm_model[vcm_model['Categorization'] == 1]

tot_vcm_reps = len(vcm_reps_id)
tot_vcm_dems = len(vcm_dems_id)
```

```python
#voters both models classified
tot_valid = len(validated_users)
tot_comp = len(comp_users)
```

```python
#affiliation of validated voters
valid_rep = validated_users[validated_users['ML Model -- Categorization'] == 0]
valid_dem = validated_users[validated_users['ML Model -- Categorization'] == 1]

tot_valid_rep = len(valid_rep)
tot_valid_dem = len(valid_dem)
```

```python
#dataframe for model data analytics
data = {'Groups' : ['Total Users', 'ML -- IDed', 'ML -- Rep', 'ML -- Dem', 'VCM -- IDed', 'VCM -- Rep', 'VCM -- Dem', 'Validated', 'Validated Republicans', 'Validated Democrats'], 
        'Totals' : [tot_users, tot_ml_id, tot_ml_reps, tot_ml_dems, tot_vcm_id, tot_vcm_reps, tot_vcm_dems, tot_valid, tot_valid_rep, tot_valid_dem]}

models_df = pd.DataFrame(data, columns = ['Groups', 'Totals'])
```

# Visual Representations of Data

```python
from pandas import Series
import seaborn as sns
import matplotlib.pyplot as plt

#values to be ploted
labels = Series(models_df['Groups'])
values = Series(models_df['Totals'])
descending = models_df.sort_values('Totals', ascending = False).Groups

plt.figure(figsize = (13,5))
sns.set(font_scale = 1.2)
sns.barplot(x = values, y = labels, data = models_df, order = descending, palette = 'Oranges_r').set_title('User ID Analysis')
```

```python
#creates figure with 3 subplots (donut pie charts)
fig, (plot_model, plot_ml, plot_vcm) = plt.subplots(1, 3, figsize = (18,7))

fig.suptitle('Model Comparison', fontsize = 'xx-large', ha = 'center')

# circles for donut pie chart
donut = plt.Circle( (0,0), 0.6, color = 'White')
donut_2 = plt.Circle( (0,0), 0.6, color = 'White')
donut_3 = plt.Circle( (0,0), 0.6, color = 'White')

# 1 -- % Model Comparison
labels_id = ['ML \n (' + str(tot_ml_id) + ')', 'VCM \n (' + str(tot_vcm_id) + ')']
values_id = [tot_ml_id, tot_vcm_id]
colors_id = ['bisque', 'orange']

plot_model.pie(values_id, labels = labels_id, colors = colors_id)
plot_model.set_title('Voters IDed', fontsize = 'large', pad = 20)
plot_model.add_artist(donut)
plot_model.text(x= -.2, y = 0, s = str(tot_users), fontsize = 'large')


# 2 -- % ML Model 
labels_aff = ['Republicans \n (' + str(tot_ml_reps) + ')', 'Democrats \n (' + str(tot_ml_dems) + ')']
size_aff = [tot_ml_reps, tot_ml_dems]
colors_aff = ['salmon', 'cornflowerblue']
per = (tot_ml_dems / tot_ml_id) * 100
per = '{0:.2f}'.format(per)

plot_ml.pie(size_aff, labels = labels_aff, colors = colors_aff)
plot_ml.set_title('ML -- % Democrats IDed', fontsize = 'large', pad = 20)
plot_ml.add_artist(donut_2)
plot_ml.text(x= -.2, y = 0, s = str(per) + "%", fontsize = 'large')


# 2 -- % VCM Model 
labels_aff = ['Republicans \n (' + str(tot_vcm_reps) + ')', 'Democrats \n (' + str(tot_vcm_dems) + ')']
size_aff = [tot_vcm_reps, tot_vcm_dems]
colors_aff = ['salmon', 'cornflowerblue']
per = (tot_vcm_dems / tot_vcm_id) * 100
per = '{0:.2f}'.format(per)

plot_vcm.pie(size_aff, labels = labels_aff, colors = colors_aff)
plot_vcm.set_title('VCM -- % Democrats IDed', fontsize = 'large', pad = 20)
plot_vcm.add_artist(donut_3)
plot_vcm.text(x= -.2, y = 0, s = str(per) + "%", fontsize = 'large')
```
