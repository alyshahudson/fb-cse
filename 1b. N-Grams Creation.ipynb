# Import Packages

```python
import pandas as pd
import re
import nltk

nltk.download('stopwords')
nltk.download('punkt') 
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.corpus import stopwords
```

# Set Parameter

```python
#frequency n-grams must at least have to be valid rules
freq_cutoff = 10
```

# Read in Voter Data

```python
input_df = pd.read_csv('Unlabeled Data.csv')
```

# Clean Up Messages

```python
#lowercase
input_df['Message'] = input_df['Message'].apply(lambda x: x.lower())

#removes all nan occurences / empty strings
input_df['Message'] = input_df['Message'].apply(lambda x: x.replace('nan', ''))

#eliminates punctuations, symbols, and whitespaces
input_df['Message'] = input_df['Message'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))
```

# Find N-Grams

```python
stopwords = stopwords.words('english')
```

```python
ngram_df = pd.DataFrame.from_dict(list(ngram_dict.items()))
ngram_df.columns = ['Terms', 'Frequency']
```
